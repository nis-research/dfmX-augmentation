```python
from operator import mod
import pickle
import matplotlib.pyplot as plt 
import numpy as np
dir = './'
all = {}
count = 0

imagenet_classes = ['airplane','automobile','bird','cat','deer','dog','frog','horse','ship','truck']

for m_path in (['resnet18_6']): #'alex',
    with open(dir+m_path+'.pkl', 'rb') as f:
        alexd = pickle.load(f)
        all.update({count:alexd})
    count += 1

name = ['ResNet9'] #'AlexNet','VGG11',
fig, axs = plt.subplots(len(name),10,sharex=True,sharey=True)
fig.set_figheight(15)
fig.set_figwidth(15)
r=0
for model in range(len(name)):

    all_mask = all[model]
    for mask_i in range(10):
        map = np.array(all_mask[mask_i])
        if model == 4:
            print(len(map[map==1]))
        axs[mask_i].imshow(map,cmap='gray')
     
    axs[r,0].set_ylabel(name[model])
    
    r += 1
plt.subplots_adjust(left=0.2, bottom=0.1, right=0.8, top=0.8, wspace=0.05, hspace=-0.85)
# plt.savefig('masks_degrasame.png',bbox_inches='tight')
plt.rcParams.update({'font.size': 10})
plt.savefig('masks_degrasame.pdf',bbox_inches='tight')
# CIFAR
```
